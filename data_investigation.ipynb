```python
import pandas as pd
import os
```

```python
data_dir = './data'

total_rows = 0
files = ['Health_Facility_General_Information.csv',
'Hospital_Inpatient_Discharges__SPARCS_De-Identified___2010.csv',
'Hospital-Acquired_Infections__Beginning_2008.csv', 'apr_drg_codes.json']
for file in files:
        if '.csv' in file:
                f = pd.read_csv(os.path.join(data_dir, file))
        else:
                f = pd.read_json(os.path.join(data_dir, file))
        f.head()
        # f.to_json('./data/apr_drg_codes.json')
        file_rows = len(f)
        print(file, file_rows)
        total_rows += file_rows

"Total rows:", total_rows
```
